Short hand notation

Merge, Join, and Concatenate Dataframes.

Just like in SQL we performs joints here also we can Combine multiple data frames or we can merge multiple data frame based on one or more common columns.
For combining this data we basically use 4 ways:
1. inner (intersection of both DF)
2. outer (union of both DF)
3. left
4. right

```python
#### Let's get started by Creating our two basic dataframe. df1 and df2
### Here we use merge, we have 4 values to pass in merge 1.left dataframe 2.right dataframe 3.how do we want to merge 4.on what we want to merge
### Like here df1 is left dataframe, df2 is right dataframe, how=inner, and in id basics we want to merge.

df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'Names':['Kajal','Sneha','Senjuti','Youtub']
})
df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'Course':['Ml','CPP','LLD','HLD']
})
inner_result=pd.merge(df1,df2, how='inner', on='id')
inner_result
```

```python
### For outer
### We get all the data from both dataframe but with some NaN (Not a Number), which is used for any missing or undefine values.Doesn't have
### to be a number every time.

outer_result=pd.merge(df1,df2, how='outer', on='id')
outer_result
```

```python
### For Left
### Give all the left values which is not null
### Right value can be null but not the left one.

left_result=pd.merge(df1,df2, how='left', on='id') 
left_result
```

```python
### For Right
### Give all the Right values which is not null
### Left value can be null but not the Right one.

right_result=pd.merge(df1,df2, how='right', on='id')
right_result
```

Right now in the above example we have the common column 'id', basically we have the same labels in both of the dataframes But that might not be a scenario in real life case. 
Bcoz in real life scenario in one data frame or csv file it might be called 'user ID' and in another csv file or data frame it might be called 'student ID' or 'Employee ID'. so basically we have same columns but differents labels in both dataframe.
And that's why for merging this two data frame we use 'right_on' or 'left_on' insted of 'on'.
Let's get started.


```python
df3=pd.DataFrame({
    'id1':[1,2,3,4],
    'Names':['Kajal','Sneha','Senjuti','Youtub']
})
df4 = pd.DataFrame({
    'id2':[3,4,5,6],
    'Course':['Ml','CPP','LLD','HLD']
})
inner_result=pd.merge(df3,df4, how='inner', left_on='id1', right_on='id2')
inner_result
```

Concatenating DataFrame (basically different ways of combining 2 Data Frame)

sometimes we want to combine our data frames but not on basicis of some columns, we just want to stack them on top of each other, that can be row wise or column wise. Basically stack them either vertically or horizontally. This is called concatenating two data frame.

```python
### here we can see that  
### here we can see that, we have to give 1.iterables either series or data frame. Basically what all the data frames we want to combine 
#                                         2.axis means either we want to do it vertically or horizontally.


df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'Names':['Kajal','Sneha','Senjuti','Youtub']
})
df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'Course':['Ml','CPP','LLD','HLD']
})
result=pd.concat([df1,df2], axis=0)             ### here we get index 0 1 2 3 0 1 2 3 bcoz we combine it as a stack but we want index
result                                          ### like 0 1 2 3 4 5 6 7 For that we gonna use ignor_index.                  
```

```python
df1=pd.DataFrame({
    'id':[1,2,3,4,],
    'Name':['Kajal','Sneha','Senjuti','Youtub']
})
df2=pd.DataFrame({
    'id':[3,4,5,6],
    'Course':['ML','CPP','LLD','HLD']
})
result=pd.concat([df1,df2], axis=0, ignore_index=True)
result
```

```python
### Horizontally

result2=pd.concat([df1,df2],axis=1)
result
```

```python
### concatinating by using keys
## here we see hierarchial indexing/multi-level indexing is created.

result3=pd.concat([df1,df2], axis=0, keys=['df1','df2'])    ## we can differentiate the row that are coming from both the Dataframe.  
result3
```
